```python
#from IPython.display import clear_output

tarefas = []

def main():

  print("Bem-vindo ao Sistema de Lista de Compras.")
  while True:
    #clear_output(wait=True)
    print("\nMenu: ")
    print("1. Adicionar item")
    print("2. Listar itens")
    print("3. Marcar item como comprado")
    print("4. Remover item")
    print("5. Editar item")
    print("6. Listar os não comprados")
    print("7. Sair")

    escolha = input("Digite o número que corresponde a opção desejada: ")
    #input("\nPressione Enter para continuar")

    if escolha == '1':
      descricao = input("Digite o nome do item: ").strip()
      if descricao == '':
        print('Nome não pode ser vazio!')
        continue
      try:
        quantidade = int(input(f"Digite a quantidade de {descricao}: "))
        if quantidade <= 0:
          print('Quantidade deve ser maior que zero!')
          continue
        item = {"item": descricao,"quantidade": quantidade ,"comprada": False}
        adicionar_tarefa(item)
      except ValueError:
        print('Entrada inválida! Por favor digite um número em quantidade')
      #input("\nPressione Enter para continuar")
    elif escolha == '2':
      visualizar_tarefas()
    elif escolha == '3':
      visualizar_tarefas()
      try:
        if len(tarefas) == 0:
          continue
        indice = int(input("Digite o número do item que você deseja marcar como comprado: ")) -1
        marcar_concluida(indice)
      except ValueError:
        print("Entrada inválida! Por favor digite um número: ")
    elif escolha == '4':
      visualizar_tarefas()
      try:
        if len(tarefas) == 0:
          continue
        indice = int(input("Digite o número do item que você quer remover: ")) -1
        remover_tarefa(indice)
      except ValueError:
        print("Entrada inválida! Por favor digite um número: ")
    elif escolha == '5':
      visualizar_tarefas()
      try:
        if len(tarefas) == 0:
          continue
        indice = int(input("Digite o número do item que você quer editar: ")) -1
        quantidde = int(input(f"Digite a nova quantidade do item {tarefas[indice]['item'].upper()}: "))
        editar_item(indice, quantidade)
      except ValueError:
        print("Entrada inválida! Por favor digite um número")
    elif escolha == '6':
      visualizar_nao_comprados()
    elif escolha == '7':
      print("Saindo do Sistema! Até logo!")
      break
    else:
      print("Opção inválida. Por favor, tente novamente")

def adicionar_tarefa(item):
  tarefas.append(item)
  print(f'Item: "{item['item']}" adicionada com sucesso!')

def visualizar_nao_comprados():
  if not tarefas:
    print('Nenhum item foi encontrado')
    return
  print('\nLista de Itens não comprados')
  for i, tarefa in enumerate(tarefas):
    if not tarefa['comprada']:
      print(f'-> [ ] {tarefa['item']} - Quantidade: {tarefa['quantidade']}')

def visualizar_tarefas():
  if not tarefas:
    print('Nenhum item foi encontrado')
    return
  print("\nLista de Itens: ")
  for i, tarefa in enumerate(tarefas):
    status = '[X]' if tarefa["comprada"] else '[ ]'
    print(f'{i + 1}. {status} {tarefa['item']} - Quantidade: {tarefa['quantidade']}')

def editar_item(indice, quantidade):
  if 0 <= indice <= len(tarefas):
    tarefas[indice]['quantidade'] = quantidade
    print(f'Item {tarefas[indice]['item']} editada com sucesso')
  else:
    print('Indice de item inválido')

def marcar_concluida(indice):
  if len(tarefas) == 0:
    return

  if 0 <= indice <= len(tarefas):
    tarefas[indice]["comprada"] = True
    print(f'Item "{tarefas[indice]['item']}" foi marcada como comprada!')
  else:
    print("Índice de item inválido")

def remover_tarefa(indice):
  if len(tarefas) == 0:
    return

  if 0 <= indice <= len(tarefas):
    descricao = tarefas.pop(indice)["item"]
    print(f'Item "{descricao}" foi removida com sucesso!')
  else:
    print("Índice de Item inválido")
```

```python
main()
```
